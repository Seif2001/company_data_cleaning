```python
def clean_and_capitalize(name):
    # Remove dashes and special characters
    name = re.sub(r'[^a-zA-Z0-9\s]', '', name)
    
    name = re.sub(r'www\.|\.com', '', name)  # Clean the company name
    name_words = name.split()
    short_words = ['cc', 'llc', 'll']  # List of words to remove
    
    # Keywords to check for
    keywords = ['starbucks', 'apple', 'carrefour', 'kfc']
    
    for i, word in enumerate(name_words):
        # Check if any keyword is in the word
        for keyword in keywords:
            if keyword in word.lower():
                return keyword.capitalize()  # Return the keyword as the cleaned company name
    
    # Remove one-character words and specified short words
    name_words = [word.capitalize() for word in name_words if len(word) > 1 and word.lower() not in short_words]
    return ' '.join(name_words)  # Capitalize and join the remaining words
```

```python
headers_Get = {
        'User-Agent': 'your bot 0.1',
        'Accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,*/*;q=0.8',
        'Accept-Language': 'en-US,en;q=0.5',
        'Accept-Encoding': 'gzip, deflate',
        'DNT': '1',
        'Connection': 'keep-alive',
        'Upgrade-Insecure-Requests': '1'
    }


def google(q):
    s = requests.Session()
    q = '+'.join(q.split())
    url = 'https://www.google.com/search?q=' + q + '&ie=utf-8&oe=utf-8'
    r = s.get(url, headers=headers_Get)

    soup = BeautifulSoup(r.text, "html.parser")
    print(soup)

    output = []
    for searchWrapper in soup.find_all('h3', {'class':'r'}): #this line may change in future based on google's web page structure
        url = searchWrapper.find('a')["href"] 
        text = searchWrapper.find('a').text.strip()
        result = {'text': text, 'url': url}
        output.append(result)

    return output

def get_google_search_results(company_name):
    search_query = f"{company_name} official website"
    search_results = googlesearch.search(search_query)
    first_link = next(search_results, None)
    
    return first_link
```
